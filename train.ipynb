# Training

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
seed = 0
```

```python
def evaluate(classifier, X, y, type='training'):
    y_hat = classifier.predict(X)
    print(f'F1 score for {type} data is :  {f1_score(y_true=y, y_pred=y_hat)}')
    print(f'Accuracy score for {type} data is :  {accuracy_score(y_true=y, y_pred=y_hat)}')
    cm = confusion_matrix(y_true=y, y_pred=y_hat, labels=[0,1])
    cmn = cm/cm.sum(axis=1)[:, np.newaxis]
    fig, ax = plt.subplots(figsize=(10,10))
    sns.heatmap(cmn, annot=True, fmt='.2f', xticklabels=[0,1], yticklabels=[0,1])
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    ax.set_title(f'Confusion matrix for {type} data');
    plt.show(block=False)
```

```python
classifiers_dict = {'LogisticRegression':(LogisticRegression(
    random_state=seed),{'LogisticRegression__C' : [0.001,0.01,0.1,10,100],
                    'LogisticRegression__penalty' : ['l2', 'l1']
                    }),
 'SVC': (SVC(random_state=seed), {'SVC__C' : [0.001,0.01,0.1,10,100],
                    'SVC__kernel' : ['linear', 'poly', 'rbf'], 
                    'SVC__degree' : (1,2,3)
                    }),
 'DecisionTreeClassifier': (tree.DecisionTreeClassifier(random_state=seed), {'DecisionTreeClassifier__max_depth' : [2,3,4,5,6],
                                                                            'DecisionTreeClassifier__min_samples_split': [2,6,8,10,15],
                                                                            'DecisionTreeClassifier__min_samples_leaf' : [2,4,8],
                                                                            'DecisionTreeClassifier__max_features' : [2,3,4,5,6,7,8]} ),
 'RandomForestClassifier':(RandomForestClassifier(random_state=seed),{'RandomForestClassifier__max_depth' : [2,3,4,5,6],
                                                                            'RandomForestClassifier__min_samples_split': [2,6,8,10,15],
                                                                            'RandomForestClassifier__min_samples_leaf' : [2,4,8],
                                                                            'RandomForestClassifier__max_features' : [2,8],
                                                                                   }),
 'GradientBoostingClassifier': (GradientBoostingClassifier(random_state=seed),{'GradientBoostingClassifier__max_depth' : [2,3,4,5,6],
                                                                            'GradientBoostingClassifier__min_samples_split': [2,6,8,10,15],
                                                                            'GradientBoostingClassifier__min_samples_leaf' : [2,4,6,8],
                                                                            'GradientBoostingClassifier__max_features' : [2,8],
                                                                            
                                                                           })
}
```

```python
def call_grid_search(clf, params, data_X,data_y,scoring,cv):
    kfolds = StratifiedKFold(cv)
    gscv = GridSearchCV(clf,param_grid=params, cv=kfolds.split(data_X,data_y), scoring=scoring,n_jobs=-1)
    gscv.fit(data_X,data_y)
    return gscv
    
def choose_best_classifier(classifiers_dict,data_X,data_y,scoring='precision',cv=5):
    best_score = 0
    best_clf = None
    for name, clf in classifiers_dict.items():
        clf, params = clf
        clf = Pipeline(steps=[('scaler',StandardScaler()),(name,clf)])
        gscv = call_grid_search(clf, params, data_X,data_y,scoring,cv)
        score = gscv.best_score_
        print('Classifier : {0}, mean {1} : {2}'.format(name,scoring, score))
        if score>best_score:
            best_score = score
            best_clf = (name, gscv.best_estimator_)
    print('Best classifier : {0}, mean {1} : {2} '.format(best_clf[0],scoring,best_score))
    return best_clf
```

```python
clf = choose_best_classifier(classifiers_dict, X_train, y_train, 'f1')
```

```python
evaluate(clf[1], X_test, y_test, 'test')
```
